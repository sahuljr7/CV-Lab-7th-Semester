```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
# Load images
query_image = cv2.imread('/content/river.jpg', cv2.IMREAD_GRAYSCALE)
train_image = cv2.imread('/content/test.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# Initialize keypoint detectors and descriptors
orb = cv2.ORB_create()
cv2.SIFT_create()
```

```python
# Detect keypoints and compute descriptors
query_kp, query_desc = orb.detectAndCompute(query_image, None)
train_kp, train_desc = orb.detectAndCompute(train_image, None)
```

```python
# Initialize brute force matcher
bf = cv2.BFMatcher()
```

```python
# Match descriptors using brute force
matches = bf.match(query_desc, train_desc)
```

```python
# Sort matches by distance
matches = sorted(matches, key=lambda x: x.distance)
```

```python
# Draw top 'n' matches
n = 10
result_image = cv2.drawMatches(query_image, query_kp, train_image, train_kp, matches[:n], None, flags=2)
```

```python
# Display result
cv2_imshow(result_image)
```
